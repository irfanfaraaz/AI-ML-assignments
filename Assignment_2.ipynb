```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Loading/reading the dataset

```python
df = pd.read_csv('House Price India.csv')
```

univariate analysis

```python
df.hist(figsize=(20,20))
plt.show()
```

 bivariate analysis by plotting scatter plots between the target variable (Price) and other variables:

```python
sns.scatterplot(x='living area', y='Price', data=df)
plt.show()
```

multivariate analysis by plotting a correlation matrix

```python
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

descriptive statistics using the describe() function:

```python
df.describe()
```

Handle missing values using the fillna() function:

```python
df.fillna(df.mean(), inplace=True)
```
